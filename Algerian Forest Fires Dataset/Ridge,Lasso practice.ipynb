```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
```

```python
dataset = pd.read_csv('Algerian_forest_fires_dataset.csv')
display(dataset)
```

```python
dataset.info()
```

```python
#Data cleaning 
dataset.isnull().sum(axis=0)
```

```python
dataset.isnull().any(axis=1)
```

```python
dataset[dataset.isnull().any(axis=1)]
```

```python
dataset.loc[:122 , 'Region']=0
dataset.loc[122: , 'Region']=1
df=dataset
```

```python
df.info()
```

```python
df[['Region']]=df[['Region']].astype(int)
```

```python
df.head()
```

```python
# removing the null values 
df = df.dropna().reset_index(drop=True)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
# remove 122 th row and update the dataframe df
df = df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
# fix spaces in columns names
df.columns = df.columns.str.strip()
df.columns
```

```python
## Changing the required  columns as integer datatypes
df[['month','day','year','Temperature','RH','Ws']] = df[['month','day','year','Temperature','RH','Ws']].astype(int)
```

```python
df.info()
```

```python
object_columns = []
for col in df.columns:
    if df[col].dtypes == 'O':
        object_columns.append(col)


object_columns
```

```python
# changing datatype to float of all feature where datatype == obkect
for i in object_columns:
    if i!='Classes':
        df[i] = df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
display(df)
```

```python
# Let save the cleaned dataset
df.to_csv('Algerian_forest_fires_cleaned_dataset.csv', index=False)
```

```python
# Deleting day ,month and year columns
df_new = df.drop(['day','month','year'],axis=1)
```

```python
df_new
```

```python
# Encoding of the categories in Classes (replacing categorcal values)
## np.where(condition, value_if_true, value_if_false)

df_new['Classes'] = np.where(df_new['Classes'].str.contains('not fire'),0,1)
df_new
```

```python
df_new['Classes'].value_counts()
```

```python
## Plot density plot for all features
plt.style.use('seaborn-v0_8')
df_new.hist(bins=50 , figsize=(20,15))
plt.show()
```

```python
## Percenatge for pie Chart
percentage = df_new['Classes'].value_counts(normalize = True) * 100

percentage
```

#### here 1 means 'fire' and 0 means 'not fire'

```python
# Pie chart-এ দেখানোর জন্য লেবেল তৈরি করা হচ্ছে
Classlabels = ['Fire', 'Not Fire']

# Figure এর আকার নির্ধারণ করা হচ্ছে (চওড়া 12 ইঞ্চি, লম্বা 7 ইঞ্চি)
plt.figure(figsize=(12, 7))

# Pie chart আঁকা হচ্ছে
# - percentage: প্রতিটি অংশের শতাংশ মানের জন্য
# - labels: প্রতিটি ভাগের নাম (Fire / Not Fire)
# - autopct: প্রতিটি ভাগে শতাংশ কত তা দেখানোর ফরম্যাট ('%1.1f%%' → যেমন: 45.0%)
plt.pie(percentage, labels=Classlabels, autopct='%1.1f%%')

# Pie chart-এর শিরোনাম সেট করা হচ্ছে
plt.title("Pie Chart for Classes")

# চূড়ান্তভাবে চার্টটি দেখানো হচ্ছে
plt.show()
```
