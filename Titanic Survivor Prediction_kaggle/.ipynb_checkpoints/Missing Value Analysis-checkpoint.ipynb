```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
train_pd=pd.read_csv('train.csv')
test_pd=pd.read_csv('test.csv')
display(train_pd)
```

```python
# Extract the 'Survived' column from the DataFrame `train_pd` and store it separately
survived = train_pd['Survived']

# Remove the 'Survived' column from `train_pd` so it no longer exists in the features DataFrame
train_pd = train_pd.drop(['Survived'], axis=1)

# Display the updated DataFrame without the 'Survived' column
display(train_pd)
```

```python
# Extract 'PassengerId' column from train and test DataFrames as indices
train_idx = train_pd['PassengerId']
test_idx = test_pd['PassengerId']

# Combine the train and test DataFrames vertically (stack rows),
# then reset the index to have a continuous range starting from 0
combined_pd = pd.concat([train_pd, test_pd]).reset_index(drop=True)

# Display the combined DataFrame
display(combined_pd)
```

```python
print("NULL VALUES :")
print(combined_pd.isnull().sum())
```

```python
# combined_pd ডেটাফ্রেমের 'Age' কলামের প্রতিটি ভ্যালুর counts সংগ্রহ করে y তে রাখছি
y = combined_pd['Age'].value_counts()

# y থেকে সব ইউনিক এজের মানগুলো নিয়ে x তে রাখছি
x = y.index.values

# প্লটের আকার নির্ধারণ করছি (দৈর্ঘ্য 10, উচ্চতা 5)
plt.figure(figsize=(10,5))

# বার চার্টে x (বয়সের মান) এবং y (প্রতিটি বয়সের সংখ্যা) প্রদর্শন করছি
plt.bar(x, y)

# চার্টের শিরোনাম সেট করছি
plt.title("All Ages")

# চার্ট দেখাচ্ছি
plt.show()
```

```python
class_age = combined_pd[['Pclass', 'Age']]
class_age
```

```python
# 'Pclass' এবং 'Age' কলাম দুইটি নিয়ে একটি নতুন DataFrame তৈরি করছি
class_age = combined_pd[['Pclass', 'Age']]

# Pclass এর মান 1 থেকে 3 পর্যন্ত প্রতিটি ক্লাসের জন্য আলাদা আলাদা বার চার্ট তৈরি করছি
for i in range(1, 4):
    plt.figure(figsize=(10, 5))               # চার্টের সাইজ নির্ধারণ
    plt.title("Class " + str(i))               # চার্টের শিরোনাম, যেমন "Class 1"
    
    # ওই ক্লাসে (Pclass == i) থাকা এজের মান গুলো গোনা
    y = class_age[class_age['Pclass'] == i]['Age'].value_counts()
    
    # গোনার রেজাল্ট থেকে বয়সের ইউনিক ভ্যালুগুলো নেওয়া
    x = y.index.values
    
    # বার চার্টে বয়স ও ফ্রিকোয়েন্সি দেখানো
    plt.bar(x, y)
    
    # চার্ট প্রদর্শন করা
    plt.show()
```

```python
class_age = combined_pd[['Pclass', 'Age']]
class_age
```

```python
# 'class_age' DataFrame-কে 'Pclass' অনুযায়ী গ্রুপ করে প্রতিটি গ্রুপের মধ্যম (median) বয়স বের করছি
medians = class_age.groupby('Pclass').median()

# গ্রুপবাই করে পাওয়া মধ্যম মানগুলো প্রিন্ট করছি
print(medians)
```

```python
for i in range(3):
    idx = np.where( (combined_pd['Pclass']== i+1 ) & (combined_pd['Age'].isnull()) )[0]
    combined_pd.loc[idx , 'Age'] = medians.values[i][0]
display(combined_pd)    
    
    
```

```python
# combined_pd ডেটাফ্রেমের 'Fare' কলামে কোন সেলে null (missing) আছে কি না তা Boolean Series হিসেবে ans এ রাখছি
ans = combined_pd['Fare'].isnull()

# যেসব সারিতে 'Fare' এর মান null, সেগুলোকে দেখাচ্ছি
display(combined_pd[ans])
```
