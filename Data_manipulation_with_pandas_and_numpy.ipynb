```python

## rename Columns product_price to room_price
df = df.rename(columns={'product_price' : 'room_price'})
df
```

```python
## based on bedrooms aggregating room price values with mean
grouped_mean = df.groupby('bedrooms')['room_price'].mean()
print(grouped_mean)
```

```python
## based on furnishing status aggregating room price values with mean
grouped_mean = df.groupby('furnishingstatus')['room_price'].mean()
print(grouped_mean)
```

```python
## based on furnishing status,bedrooms,bathrooms aggregating room price values with sum
grouped_mean = df.groupby(['furnishingstatus','bedrooms','bathrooms'])['room_price'].sum()
print(grouped_mean)
```

```python
## aggregate multiple functions like mean sum count
grouped_agg = df.groupby('furnishingstatus')['room_price'].agg(['mean','sum','count'])
grouped_agg
```

```python
## Merging and joining Dataframes
# creating Dataframes
df1=pd.DataFrame({'Key': ['A','B','C'] , 'value1':[1,2,3]})
df2=pd.DataFrame({'Key': ['A','B','D'] , 'value1':[4,5,6]})
df1
```

```python
df2
```

```python
## Merge DataFrames
# Perform an inner join between df1 and df2 on the 'Key' column.
# Only the rows with matching 'Key' values in both DataFrames will be included in the result.

pd.merge(df1,df2 , on ='Key', how = 'inner')
```

```python
## Merge DataFrames
# Perform an outer join between df1 and df2 on the 'Key' column.
# Only the rows with all the 'Key' values in both DataFrames will be included in the result.

pd.merge(df1,df2 , on ='Key', how = 'outer')
```
