```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5,6,7]
y=[1,4,9,16,25,36,49]
##create a line plot 
plt.plot(x,y)
plt.xlabel("X axis ")
plt.ylabel("Y label ")
plt.title(" Basic Line PLot")
plt.show()

```

```python
x=[1,2,3,4,5,6,7]
y=[1,4,9,16,25,36,49]

plt.plot(x,y, color='red',linestyle='-.',marker='o',linewidth=3, markersize=9 )
plt.grid(True)
```

```python
x=[1,2,3,4,5,6,7]
y1 =[1,4,9,16,25,36,49]
y2 = [ 1,2,3,4,5,6,7]
plt.figure(figsize=(9,5))
plt.subplot(1,2,1)
plt.plot(x,y1,color='green')
plt.title("Plot 1")

plt.subplot(1,2,2)
plt.plot(y1,x,color='green')
plt.title("Plot 2")
```

```python
# Bar plot 
categories = ['A','B','C','D','E']
values = [5,7,3,8,6]
plt.bar( categories , values , color ='purple')
```

```python
# Histogram Creation
data =[ 1,2,2,2,3,4,5,3,3,4,4,5,1,5,5]

plt.hist(data,bins=5, color='orange',edgecolor='black')
```

```python
# Scatter plot
x = [1 , 2, 3, 4, 5]
y = [2 , 3, 4, 5, 6]
plt.scatter(x,y,color='blue', marker='x')
```

```python
## pie chart
labels = ['A','B','C','D']
sizes = [30,20,40,10]
colors = ['gold','yellowgreen','lightcoral','lightskyblue']
explode =(0.2,0,0,0)
plt.pie(sizes, labels=labels , autopct='%1.1f%%',shadow=True , explode = explode)
```

```python
## Data Visualization
import pandas as pd
Housing_data= pd.read_csv('Housing.csv')
display(Housing_data)
```

```python
Housing_data.info()
```

```python
## Plot total price by bedrooms
total_price_by_bedrooms = Housing_data.groupby('bedrooms')['price'].sum()
print (total_price_by_bedrooms)
```

```python
total_price_by_bedrooms.plot(kind='bar',color='teal')
```

```python
## Plot total price by furnishingstatus
total_price_by_furnishingstatus = Housing_data.groupby('furnishingstatus')['price'].sum()
print (total_price_by_furnishingstatus)
```

```python
total_price_by_furnishingstatus.plot(kind='bar',color='orange')
```

```python
total_price_by_furnishingstatus.plot(kind='pie',color='orange', autopct='%1.1f%%')
```

```python
## price trend over parking
price_trend = Housing_data.groupby('parking')['price'].sum()
price_trend
```
